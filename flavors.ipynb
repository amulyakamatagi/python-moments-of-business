```python
sns.set_style(style='whitegrid')
figure=plt.figure(figsize=(7,7))
sns.barplot(x='Base Flavor',y='Flavor Rating',data=data,hue='Base Flavor')
```

```python
sns.set_style(style='whitegrid')
figure=plt.figure(figsize=(7,7))
sns.barplot(x='Base Flavor',y='Texture Rating',data=data,hue='Base Flavor')
```

```python
# texture and flavour vice , chocolate is comparitively rated higher than that of vanilla
```

```python
sns.set_style(style='whitegrid')
figure=plt.figure(figsize=(7,7))
sns.barplot(x='Liked',y='Flavor Rating',data=data,hue='Base Flavor')
```

```python
# most of the customers liked chocolate over vanilla
```

```python
figure=plt.figure(figsize=(10,9))
sns.barplot(x='Total Rating',y='Flavor',data=data,hue='Flavor')
```

```python
# from the above visualization , we can comprehend that 'Pistachio'is least flavoured ice-cream (>7.5)
# mint chocolate chip , with base flavour Vanilla is top rated
```

# **UNDERSTANDING THE RELATION BETWEEN THE VARIABLES THROUGH SCATTERPLOT**

```python
# scatterplot b/w texture_rating and total_rating
```

```python
figure=plt.figure(figsize=(8,9))
sns.scatterplot(x='Texture Rating',y='Total Rating',data=data,color='darkred',s=100)
plt.title('TEXTURE_RATING/TOTAL_RATING')
```

```python
# the variables are positively_linear_associated with each other
```

```python
figure=plt.figure(figsize=(8,9))
sns.scatterplot(x='Flavor Rating',y='Texture Rating',data=data,color='crimson',s=100)
plt.title('FLAVOR_RATING/TEXTURE_RATING')
```

```python
# the variables are positively linear associated with eachother
```
