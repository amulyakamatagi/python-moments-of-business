# **PIE PLOT**
(calculating the praportion of rank(prof , assoc prof and assistent prof , gender (male and female) & discipline (A&B)))

```python
# praporting of assistant professors , professors and associate professors
```

```python
rank_count=df["rank"].value_counts()
per=(rank_count/77*100)
per
```

```python
sex_count=df["sex"].value_counts()
per_1=(sex_count/77*100)
per_1
```

```python
dis_count=df["discipline"].value_counts()
per_2=(dis_count/77*100)
per_2
```

```python
plt.subplot(2,2,1)
per.plot(figsize=(10,8),kind="pie")
plt.subplot(2,2,2)
per_1.plot(figsize=(10,8),kind="pie")
plt.plot(2,2,3)
per_2.plot(kind="pie")
```

# **INTERPRETATION**

* the praportion of professors are comparitively more than assistent professor and associate professor
* while Male  with B Discipline is slightly  more than Female with A discipline


# **UNIVARIATE ANALYSIS** (analysing salary across different columns)
salary (professor  vs  assocoate professor  vs  assistent professor )


```python
# salary vs rank
# which rank employee  is earning more
```

```python
sns.set_style(style="darkgrid")
sns.displot(df,x="salary",col="rank",kde=True,hue="rank")
sns.displot(df,x="salary",col="discipline",kde=True,hue="discipline")
sns.displot(df,x="salary",col="sex",kde=True,hue="sex")
```

# **INTERPRETATION**
* from the above visualizations , we can make the following interpretations
* among the ranks (professor,  associate professor  and assistent professor ) professors are earning comparitively more than any other rank
* among disciplines , B discipline is earning more than A discipline
*


```python

sns.displot(df,x="salary",hue="sex",kind="kde")
plt.figure(figsize=(10,8))
sns.displot(df,x="salary",hue="rank",kind="kde")
sns.displot(df,x="salary",hue="discipline",kind="kde")
```
