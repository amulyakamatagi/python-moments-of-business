**Hyplothesis Testing**

# Hypothesis Testing

 Ho=no difference in the TAT reports of 4 laboratories

 H1= there is a difference in TAT reports of 4 laboratories

```python
# since we have to compare average TAT reports of more than 2 samples
# ANOVA (f-statistical)test is suitable
```

```python
stats.f_oneway(lab['Laboratory 1'],lab['Laboratory 2'],lab['Laboratory 3'],lab['Laboratory 4'])
```

```python
p_value=2.1156708949992414e-57
```

```python
if p_value>0.05:
  print('Accept Null hypothesis')
else:
  print('Accept Alternate Hypothesis')
```

**CONCLUSION :**  

the TAT reports is different for 4 laboratories
