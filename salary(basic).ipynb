```python
data=df.set_index(['sex','rank'])
```

```python
data.loc['Male']
```

```python
# visualizing
```

```python
sns.set_style(style='whitegrid')
figure=plt.figure(figsize=(8,8))
sns.histplot(x='sex',y='salary',hue='sex',data=df)
```

From the above plot, we can understand that Male  are comparitively earning higher(>1,80,000) than that of Female
(>1,60,000)

```python
sns.set_style(style='whitegrid')
figure=plt.figure(figsize=(8,8))
sns.barplot(x='rank',y='salary',data=df,hue='rank')
```

# **INFERENCE:**
from the above visualization , we can comprehend ranks who are receiving higher salary
* among all the ranks , the salary of an Assistent professor is low while a professor's salary is higher
* while an associate professor's salary is  comparitively lower than an profreesor and higher than Assistant professor  

```python
sns.set_style(style='whitegrid')
figure=plt.figure(figsize=(8,8))
sns.lineplot(x='rank',y='salary',data=df)
```

```python
sns.set_style(style='whitegrid')
figure=plt.figure(figsize=(8,8))
sns.lineplot(x='sex',y='salary',data=df)
```

```python
# comparitively , Male is getting high paid compared to female
```

```python
# groupby
```

```python
data.groupby(['rank','sex']).agg({'salary':['mean','max','min']})
```

# **INFERENCE:**

* through the above function , we can analyse the mean salary of different ranks and gender

```python
figure=plt.figure(figsize=(9,9))
heatmap=sns.heatmap(data.corr(),cmap='Blues',vmin=-1,vmax=1,annot=True)
heatmap.set_title('correlation_heatmap')
```

```python
# scatterplot
```

```python
sns.scatterplot(x='service',y='salary',data=data,color='teal')
plt.title('SCATTERPLOT')
```

from the scatterplot , we can comprehend that the variables are positively linear associated to each other
as the no of years served increases , the salary also increases

```python
sns.scatterplot(x='phd',y='salary',data=data,color='darkred')
```
