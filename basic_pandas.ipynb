```python
# checking for null values
```

```python
# boxplots
```

```python
sns.boxplot(x='salary',y='sex',data=data)
```

```python
# understanding the relation using scatterplot
```

```python
figure=plt.figure(figsize=(9,5))
sns.scatterplot(x='phd',y='salary',data=df,color='green')
```

```python
# positive linear association
```

```python
figure=plt.figure(figsize=(9,5))
sns.scatterplot(x='service',y='salary',data=df,color='black')
```

```python
# as the no of yeras served increases , salary also increases
```

```python
figure=plt.figure(figsize=(9,5))
sns.histplot(x='rank',y='salary',data=df,color='red')
```

```python
# the employee with professor rank is receiving highest salary (>1,80,000)
# while the employee with assistent professor is receiving the lowest salary(>1,00,000)
```

```python
# comparing the salaries among different disciplines
```

```python
sns.barplot(x='discipline',y='salary',data=df,hue='discipline')
```

```python
# compared to 'A'discipline, 'B' discipline is receiving more salary(>1,00,000)
```

```python
# comparing the salaries of genders
```

```python
sns.barplot(x='sex',y='salary',data=df,hue='sex')
```

```python
# male is getting the higher salary compared to female
```
