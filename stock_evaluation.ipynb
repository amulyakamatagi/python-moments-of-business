```python
# importing all libraries
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
hdfc_df=pd.read_csv('HDFC.csv')
hdfc_df
```

```python
icici_df=pd.read_csv('ICICI.csv')
icici_df
```

```python
hdfc_df.head()
```

```python
icici_df.head()
```

```python
hdfc=hdfc_df[['Date','Close Price']]
hdfc
```

```python
icici=icici_df[['Date','Close Price']]
icici
```

```python
# setting the date values as index
```

```python
hdfc_bank=hdfc.set_index(pd.DatetimeIndex(hdfc['Date']))
hdfc_bank
```

```python
icici_bank=icici.set_index(pd.DatetimeIndex(icici['Date']))
icici_bank
```

```python
# ploting the closing price
```

```python
plt.plot(hdfc_bank['Close Price'])
plt.xlabel('Time')
plt.ylabel('Closing price')
plt.title('HDFC_BANK')
```

```python
plt.plot(icici_bank['Close Price'])
plt.xlabel('Time')
plt.ylabel('Closing price')
plt.title('ICICI_BANK')
```

```python
# calculate the gains
```

```python
hdfc_bank['gain']=hdfc_bank['Close Price'].pct_change(periods=1)
hdfc_bank
```

```python
icici_bank['gain']=icici_bank['Close Price'].pct_change(periods=1)
icici_bank
```

```python
# dropping the null values
```

```python
hdfc_bank=hdfc_bank.dropna()
icici_bank=icici_bank.dropna()
```

```python
hdfc_bank
```

```python
icici_bank
```

```python
# checking for null values
```

```python
hdfc_bank.isnull().sum()
```

```python
icici_bank.isnull().sum()
```

```python
# hence the data is free from null values
```

```python
# plotting the returns
```

```python
plt.plot(hdfc_bank.index,hdfc_bank['gain'])
plt.xlabel('Time')
plt.ylabel('Gain')
plt.title('HDFC GAINS')
```

```python
plt.plot(icici_bank.index,icici_bank['gain'])
plt.xlabel('Time')
plt.ylabel('Gain')
plt.show()
```

```python
# calculating risk of each bank
```

```python
hdfc_bank['gain'].std()
```

```python
icici_bank['gain'].std()
```

```python
hdfc_bank['gain'].var()
```

```python
icici_bank['gain'].var()
```

```python
# distribution plot
```

```python
X=pd.DataFrame(columns=['hdfc_bank.gain','icici_bank.gain'])
X['hdfc_bank.gain']=pd.Series(hdfc_bank.gain)
X['icici_bank.gain']=pd.Series(icici_bank.gain)
sns.distplot(X['hdfc_bank.gain'],label='HDFC')
sns.distplot(X['icici_bank.gain'],label='ICICI')
plt.xlabel('GAIN')
plt.ylabel('DENSITY')
plt.legend()
```

```python
# probability of making 2% loss in hdfc
```

```python
from scipy import stats
stats.norm.cdf(-0.02,
hdfc_bank.gain.mean(),
hdfc_bank.gain.std())
```

```python
# probability of making 2% gain in hdfc
```

```python
1-stats.norm.cdf(0.02,
hdfc_bank.gain.mean(),
hdfc_bank.gain.std())
```

```python
# probability of making 2% loss in icici
```
