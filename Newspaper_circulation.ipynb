```python
# scatterplot(to know the direction of two variables)
```

```python
figure=plt.figure(figsize=(9,9))
sns.scatterplot(x='daily',y='sunday',data=df,color="blue")
```

```python
# there is a positive linear association between two variables
# as the daily circulation increases , sunday circulation also increases
```

```python
# correlation (to know the strength between two variables)
```

```python
figure=plt.figure(figsize=(9,9))
sns.set_style(style="whitegrid")
sns.heatmap(df_1.corr(),annot=True,cmap='crest')
```

```python
# there is a very strong positive correlation between two variables
# in implies that if x(daily)which is independent varible increases , y(sunday) which is target variable also increases
```

```python
# regression plot(to visualize the line of best fit)
```

```python
figure=plt.figure(figsize=(9,9))
sns.regplot(x='daily',y='sunday',data=df,color='green')
```

```python
from scipy import stats
import statsmodels.formula.api as smf
model=smf.ols('sunday~daily',data=df).fit()
model.summary()
```

```python
model.params # coefficients
```

```python
model.resid
```

```python
predict=model.predict()
predict
```
